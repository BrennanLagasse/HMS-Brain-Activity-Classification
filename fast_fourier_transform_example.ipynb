The goal of this file is to provide an example of the numpy build in FFT function as a sanity check

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate example time series data
# Here, let's generate a sine wave with some added noise
t = np.linspace(0, 1, 1000)  # Time points
time_series = np.sin(2 * np.pi * 5 * t) + 0.5 * np.random.randn(1000)  # Sine wave with noise

# Perform DFT using numpy's rfft function
dft_result = np.fft.rfft(time_series)

print(t[1])
print(t[0])

# Frequencies corresponding to DFT result
frequencies = np.fft.rfftfreq(len(time_series), t[1] - t[0])

# Extract amplitude and phase
amplitude = np.abs(dft_result)
phase = np.angle(dft_result)

# Original Graph
plt.figure(figsize=(10, 6))
plt.plot(t, time_series)
plt.title('Time Series Data')
plt.xlabel('Time')
plt.ylabel('Amplitude')

# Plotting Amplitude Spectrum
plt.figure(figsize=(10, 6))
plt.scatter(frequencies, amplitude, marker=".")
plt.title('Amplitude Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.xlim(0, 20)
plt.show()

# Plotting Phase Spectrum
plt.figure(figsize=(10, 6))
plt.scatter(frequencies, phase, marker=".")
plt.title('Phase Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Phase (radians)')
plt.xlim(0, 20)
plt.show()

# Reconstruct the time series using inverse FFT
reconstructed_data = np.fft.irfft(amplitude * np.exp(1j * phase))

# Plotting
plt.figure(figsize=(10, 6))

plt.plot(t, time_series, label='Original Data')
plt.plot(t, reconstructed_data, label='Reconstructed Data', linestyle='--')
plt.title('Original Data vs. Reconstructed Data')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.show()
```
